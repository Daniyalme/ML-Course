```python
import pandas as pd
import numpy as np
import os

pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
%matplotlib inline
```

### Part f)

```python
data = pd.DataFrame(
    [
        [1,2,1],
        [2,1,1],
        [2,2,0],
        [2,3,1],
        [3,1,0],
        [3,2,1],
        ],
    columns=["X1", "X2", "y"]
    )

data.head()
```

```python
plt.figure(figsize=(8,4))

sns.scatterplot(x=data["X1"], y=data["X2"], hue=data["y"])
```

```python
X = data.iloc[:,:-1]
y= data.iloc[:,-1]

h = 0.02  # Step size in the mesh
x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Create a KNN classifier with k=
knn = KNeighborsClassifier(n_neighbors=1)

# Train the classifier
knn.fit(X, y)

# Predict the class labels for each point on the meshgrid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=y, palette='Paired', edgecolor='k')
plt.contourf(xx, yy, Z, alpha=0.8, cmap='Paired')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KNN Decision Boundary')
plt.show()
```

```python
# Create a KNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier
knn.fit(X, y)

# Predict the class labels for each point on the meshgrid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=y, palette='Paired', edgecolor='k')
plt.contourf(xx, yy, Z, alpha=0.8, cmap='Paired')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KNN Decision Boundary')
plt.show()
```
